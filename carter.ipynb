## 224. Basic Calculator 

````python
from collections import deque 
def basic_calculator(s): 
    print('caulcuating') 
    sign = 1 
    res = 0 
    stack = deque() 
    i = 0 
    while i < len(s): 
#         print(i, dig) 
        dig = s[i]
        if dig.isdigit(): 
            operand = int(dig) 
            ## notice: check the next element, should be i+1 
            while i+1 < len(s) and s[i+1].isdigit(): 
                operand = operand*10 + int(s[i+1]) 
                i += 1    ## notice 
            res += sign * operand 
        elif dig == '+': 
            sign = 1 
        elif dig == '-': 
            sign = -1 
        elif dig == '(': 
            # push 
            stack.append(res) 
            stack.append(sign) 
            res = 0 
            sign = 1 
        elif dig == ')': 
            res = res * stack.pop() + stack.pop() 
        i += 1 
        
    return res  
    
    
in_str = "2+3- 999" 

res_1 = basic_calculator(in_str) 
print(res_1)
````

````python
from collections import defaultdict 
words = ["cat", "baby", "dog", "bird", "car", "ax"] 
s = "tcabnihrdjsb" 

def word_cons(words, s): 
    print('constructing') 
    res = [] 
    letter_counter = defaultdict(int) 
    for ch in s: 
        letter_counter[ch] += 1 
#     print(letter_counter)
    for word in words: 
        word_dict = defaultdict(int) 
        res.append(word)
        for ch in word: 
            word_dict[ch] += 1 
        for ch, num in word_dict.items(): 
            if ch in letter_counter: 
                if num > letter_counter[ch]: 
                    res.pop() 
                    break 
            else: 
                res.pop() 
                break 
                    
            
    
    return res 

word_cons(words, s) 
````

## 79. Word Search 
[79. Word Search](https://leetcode.com/problems/word-search/) 

````python
class Solution:
    def exist(self, board, word: str) -> bool:
        self.row = len(board) 
        self.col = len(board[0]) 
        self.board = board 
        
        for i in range(self.row): 
            for j in range(self.col): 
                if board[i][j] == word[0] and self.dfs(i, j, word, 0): 
                    return True 
        return False 
        
    
    def dfs(self, r, c, word, index): 
        if index == len(word): 
            return True 
        if r < 0 or c < 0 or r >= self.row or c >= self.col or word[index] != self.board[r][c]: 
            return False 
        
        self.board[r][c] = '#' 
        
        res = self.dfs(r-1, c, word, index+1) or self.dfs(r+1, c, word, index+1) or self.dfs(r, c-1, word, index+1) or self.dfs(r, c+1, word, index+1)
        
        self.board[r][c] = word[index] 
        
        return res 
s = "ccc" 
r = 'ere'
grid = [['c', 'c', 'x'],
        ['a', 'c', 'n'], 
        ['k', 'a', 'i'],
       ]

a = Solution() 
a.exist(grid, r)
````
