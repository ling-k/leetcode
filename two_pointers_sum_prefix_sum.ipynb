## 215. Kth Largest Element in an Array
[215. Kth Largest Element in an Array](https://leetcode.com/problems/kth-largest-element-in-an-array/)

[Quickselect Algorithm](https://www.geeksforgeeks.org/quickselect-algorithm/)

#### two pointers and three regions 
- [0, i): numbers less than pivot 
- [i, j]: unexplored numbers 
- (j, n-1]: numbers greater than or equal to pivot 


```python
class Solution:
    def findKthLargest(self, nums, k: int) -> int: 
        if len(nums) < k: 
            return 
        n = len(nums) 
        
        return self.quick_select(nums, 0, n - 1, n - k) 
        
    def quick_select(self, nums, left, right, k): 
        # base case 
        if left == right: 
            return nums[left] 
        
        pivot_idx = self.partition(nums, left, right) 
        if k == pivot_idx: 
            return nums[k] 
        elif k < pivot_idx: 
            return self.quick_select(nums, left, pivot_idx - 1, k) 
        else: 
            return self.quick_select(nums, pivot_idx + 1, right, k) 
            
        
    def partition(self, nums, l, r):  
        pivot = nums[r] 
        last = r 
        
        r -= 1 
        
        while l <= r: 
            if nums[l] < pivot: 
                l += 1 
            else: 
                nums[l], nums[r] = nums[r], nums[l] 
                r -= 1 
        nums[l], nums[last] = nums[last], nums[l]
            
        return l 

```

[75. Sort Colors](https://leetcode.com/problems/sort-colors/)

```python
class Solution:
    def sortColors(self, nums):
        """
        Do not return anything, modify nums in-place instead.
        """
        
        n = len(nums)
        i = 0 
        j = 0 
        k = n - 1 
        
        while j <= k: 
            if nums[j] == 0: 
                nums[i], nums[j] = nums[j], nums[i] 
                i += 1 
                j += 1 
            elif nums[j] == 1: 
                j += 1 
            else: 
                nums[j], nums[k] = nums[k], nums[j] 
                k -= 1 
        
```

## 1. Two Sum
[1. Two Sum](https://leetcode.com/problems/two-sum/) 

```python
class Solution:
    def twoSum(self, nums, target: int): 
        s = {} 
        n = len(nums) 
        for i in range(n): 
            cur_partner = target - nums[i] 
            if cur_partner in s: 
                return i, s[cur_partner] 
            else: 
                s[nums[i]] = i 
        return 
```

## 167. Two Sum II - Input Array Is Sorted 
[167. Two Sum II - Input Array Is Sorted](https://leetcode.com/problems/two-sum-ii-input-array-is-sorted/)

```python
class Solution:
    def twoSum(self, numbers, target: int): 
        n = len(numbers) 
        
        i = 0  
        j = n - 1   
        while i < j: 
            cur_sum = numbers[i] + numbers[j] 
            if cur_sum == target: 
                return i + 1, j + 1 
            elif cur_sum < target: 
                i += 1 
            else: 
                j -= 1 
        
        
    def twoSum_hash(self, numbers, target: int): 
        visited = {} 
        n = len(numbers) 
        for i in range(n): 
            cur_partner = target - numbers[i] 
            if cur_partner in visited: 
                return visited[cur_partner], i + 1 
            else: 
                visited[numbers[i]] = i + 1 
        return 
```

## 1099. Two Sum Less Than K
[1099. Two Sum Less Than K](https://leetcode.com/problems/two-sum-less-than-k/) 

```python
class Solution:
    def twoSumLessThanK(self, nums, k: int) -> int: 

        nums.sort() 
        res = -1 
        i = 0 
        j = len(nums) - 1 
        
        while i < j: 
            cur_sum = nums[i] + nums[j]  
            if cur_sum < k: 
                i += 1   
                res = max(res, cur_sum)
            else: 
                j -= 1 
        
        return res  
```

## 15. 3Sum
[15. 3Sum](https://leetcode.com/problems/3sum/) 

```python
class Solution:
    def threeSum(self, nums):
        if len(nums) <= 2: 
            return [] 
        
        nums.sort() 
        res = [] 
        
        n = len(nums) 
        for i in range(n): 
            if nums[i] > 0: 
                break 
            if i == 0 or nums[i - 1] != nums[i]: 
                self.two_sum(nums, i, res) 
        
        return res  
    
    def two_sum(self, nums, i, res): 
        low = i + 1 
        high = len(nums) - 1 
        target = -nums[i] 
        while low < high:
            cur_sum = nums[low] + nums[high] 
            if cur_sum == target: 
                res.append([nums[i], nums[low], nums[high]])  
                low += 1 
                high -= 1 
                while low < high and nums[low] == nums[low-1]: 
                    low += 1 
            elif cur_sum < target: 
                low += 1 
            else: 
                high -= 1 
            
```

## 16. 3Sum Closest 
[16. 3Sum Closest](https://leetcode.com/problems/3sum-closest/)

- notice: 
    - sort first  
    - if cur_sum == target, continue to search, i += 1 and j -= 1. 

```python
class Solution:
    def threeSumClosest(self, nums, target: int) -> int:
        if len(nums) <=  2: 
            return  
        nums.sort()
        n = len(nums) 
        self.gap = float('inf') 
        self.res = 0 
        for i in range(n): 
            self.two_sum(nums, i, target) 
        
        return self.res 
    
    def two_sum(self, nums, i, target): 
        low = i + 1 
        high = len(nums) - 1 
        target = target - nums[i]  
        
        while low < high: 
            cur_sum = nums[low] + nums[high] 
            cur_gap = abs(target - cur_sum) 
            cur_res = cur_sum + nums[i] 
            if cur_sum == target: 
                self.res = cur_res 
                self.gap = 0 
                low += 1 
                high -= 1 
            elif cur_sum < target:  
                if self.gap > cur_gap: 
                    self.gap = cur_gap 
                    self.res = cur_res 
                
                low += 1 
            else: 
                if self.gap > cur_gap: 
                    self.gap = cur_gap 
                    self.res = cur_res 
                high -= 1 
                
```

## 259. 3Sum Smaller 
[259. 3Sum Smaller](https://leetcode.com/problems/3sum-smaller/)

```python
class Solution:
    def threeSumSmaller(self, nums, target: int) -> int:
        if len(nums) <= 2: 
            return 0 
        
        nums.sort() 
        self.res = 0 
        n = len(nums) 
        
        for i in range(n): 
            self.two_sum(nums, i, target)
        
        return self.res 
    
    def two_sum(self, nums, i, target): 
        l = i + 1 
        r = len(nums) - 1 
        cur_fix = nums[i] 
        
        while l < r: 
            cur_sum = cur_fix + nums[l] + nums[r] 
            if cur_sum < target: 
                self.res += r - l 
                l += 1 
            else: 
                r -= 1 
            
```

## 18. 4Sum 
[18. 4Sum](https://leetcode.com/problems/4sum/)

```python
class Solution:
    def fourSum(self, nums, target: int):
        if len(nums) <= 3: 
            return [] 
        
        nums.sort() 
        self.res = [] 
        # self.target = target 
        n = len(nums)
        for i in range(n - 3): 
            if i == 0 or nums[i] != nums[i - 1]: 
                # three_l = i + 1 
                # three_r = n - 2 
                for j in range(i + 1, n - 2): 
                    if j == i + 1 or nums[j] != nums[j - 1]: 
                        self.two_sum(nums, nums[i], nums[j], j + 1, n - 1, target) 
                
        return self.res  
    
    def two_sum(self, nums, a, b, l, r, target): 
        target = target - a - b 
        while l < r: 
            cur_sum = nums[l] + nums[r] 
            if cur_sum == target: 
                self.res.append([a, b, nums[l], nums[r]]) 
                l += 1 
                r -= 1 
                while l < r and nums[l] == nums[l - 1]: 
                    l += 1 
            elif cur_sum < target: 
                l += 1  
            else: 
                r -= 1 
```

## 560.Subarray Sum Equals K  
[560.Subarray Sum Equals K ](https://leetcode.com/problems/subarray-sum-equals-k/)

```python
# brute force 
# time: O(n^3) 
# space: O(1) 
class Solution:
    def subarraySum(self, nums, k: int) -> int: 
        """
        brute force: 
        consider all possible subarrays, and sum up 
        """ 
        res = 0 
        n = len(nums)
        for i in range(n): 
            for j in range(i+1, n + 1): 
                sub_array = nums[i:j] 
                cur_sum = sum(sub_array) 
                if cur_sum == k: 
                    res += 1 
        
        return res 
```

```python
class Solution:
    def subarraySum(self, nums, k: int) -> int: 
        """
        brute force: 
        consider all possible subarrays, and sum up 
        
        Every possible indeces: 
        [5, -2, 2, -8, 1, 7, 10, 23]  
             i1     j3
        calculate a prefix sum array: 
        [0, 5,  3, 5, -3, 2, 9, 19, 42] 
        1 --> 2: cum_sum[2] - cum_sum[0] 
        i --> j: c[j+1] - c[i] 
        
        using hash map: 
        as we scan the array, we figure out all the possible results. 
        
        
        """ 
        from collections import defaultdict 
        res = 0 
        n = len(nums)  
        sum_occurrence_map = defaultdict(int) 
        sum_occurrence_map[0] = 1 

        cur = 0 
        for i in range(n):  
            cur += nums[i] 
            if cur - k in sum_occurrence_map: 
                res += sum_occurrence_map[cur - k] 
            sum_occurrence_map[cur] += 1 

            
        return res 
```

```python
class Solution:
    def checkSubarraySum(self, nums, k: int) -> bool: 
        """
        cul sum: 
        sum % k == 0 
        
        sum % k = (ps[j] - ps[i]) % k = 0 ==> 
        ps[j] % k = ps[i] % k 
        
        
        """ 
        from collections import defaultdict 
        n = len(nums) 
        sum_occur = defaultdict(int) 
        cur = 0 
        
        for i in range(n):  
            cur += nums[i] 
            cur_mod = cur % k 
            # or add to dict: 0: -1 
            if cur_mod == 0 and i != 0: 
                return True 
            if cur_mod in sum_occur: 
                if i - sum_occur[cur_mod] > 1: 
                    return True  
                    
            else: 
                sum_occur[cur_mod] = i  

            
        return False 
```

## 528. Random Pick with Weight
[528. Random Pick with Weight](https://leetcode.com/problems/random-pick-with-weight/) 

```python
class Solution:

    def __init__(self, w):
        self.w = w 
        self.n = len(w) 
        self.prefix_array = [0 for _ in range(self.n)] 
        cur = 0 
        for i in range(self.n):  
            cur += self.w[i] 
            self.prefix_array[i] = cur 
        self.sum_weight = self.prefix_array[-1] 
            
        

    def pickIndex(self) -> int:
        prob = self.sum_weight * random.random() 
        for i, item in enumerate(self.prefix_array): 
            if prob <= item: 
                return i 
```

## 238. Product of Array Except Self
[238. Product of Array Except Self](https://leetcode.com/problems/product-of-array-except-self/) 

```python
class Solution:
    def productExceptSelf(self, nums):
        """
              4  5  1  8  2 
        left  1  4  20 20 160 
        right 80  16   8  2   1
        
        
        """ 
        n = len(nums)
        left_prefix = [1 for _ in range(n)] 
        right_prefix = [1 for _ in range(n)] 
        res = [1 for _ in range(n)] 
        
        # left
        prev = 1 
        for i in range(1, n): 
            prev = prev * nums[i - 1]
            left_prefix[i] = prev 
        
        prev = 1 
        for i in range(n-2, -1, -1): 
            prev = prev * nums[i + 1] 
            right_prefix[i] = prev 
        
        for i in range(n): 
            res[i] = left_prefix[i] * right_prefix[i]

        return res 
        
```

## 370. Range Addition 
[370. Range Addition](https://leetcode.com/problems/range-addition/) 

```python
class Solution:
    def getModifiedArray(self, length: int, updates): 
        """
        [[1,2,2],[2,4,3],[0,2,-2]]
        0 0 0 0 0 
        0 2 2 2 2 
        
        0 2 2 2 2 
              -2 -2        
        
        0 2 2 2 0 (2-2) 
        
        0 2 2 2 0
        0 2 5 5 3
        
        0  2  5  5  3
       -2  0  3  3  1
       -2  0  3  5  3
        
        0  2  0  0  -2 
        
        0  2  0  0  -2 
        
        0  2  3  0  -2     
        
        0  2  3  0  -2 
        
        -2  2  3 2   -2
        
        -2  0  3 5   3  
        
        
        """
        res = [0 for _ in range(length)] 
        
        for start, end, inc in updates: 
            res[start] += inc 
            if end + 1 < length: 
                res[end+1] -= inc 
        for i in range(1, length): 
            res[i] += res[i - 1] 
            
        return res 
        
```

## 1094. Car Pooling 
[1094. Car Pooling](https://leetcode.com/problems/car-pooling/) 

```python
class Solution:
    def carPooling(self, trips, capacity: int) -> bool: 
        """
        [[2,1,5],[3,3,7]]
        
        0 0 0 0 0 0 0
        
        0 2 2 2 2 2 0  
        
        0 2 2 5 5 5 0  
        
        """ 
        time_stamp = []  
        for inc, start, end in trips: 
            time_stamp.append((start, inc)) 
            time_stamp.append((end, -inc))  
        time_stamp.sort() 

        cur_passenger = 0 
        n = len(time_stamp)
        for i in range(n): 
            cur_passenger += time_stamp[i][1] 
            if cur_passenger > capacity: 
                return False 
        
        return True 
            
```

## 1109. Corporate Flight Bookings
[1109. Corporate Flight Bookings](https://leetcode.com/problems/corporate-flight-bookings/)

```python
class Solution:
    def corpFlightBookings(self, bookings, n: int): 
        """
        [1 2 3 4 5] 
         3 3 
        
        """
        res = [0 for _ in range(n)] 
        
        for start, end, seats in bookings: 
            res[start - 1] += seats 
            if end < n: 
                res[end] += -1 * seats 
        # print(res) 
        
        for i in range(1, n): 
            res[i] += res[i - 1] 
        
        return res 
```

## 1732. Find the Highest Altitude
[1732. Find the Highest Altitude](https://leetcode.com/problems/find-the-highest-altitude/) 

```python
class Solution:
    def largestAltitude(self, gain) -> int:
        """
        [-4,-3,-2,-1,4,3,2]
        
        
        """
        n = len(gain) 
        if n == 0: 
            return 0 
        
        res = float('-inf') 
        cur = 0 
        for i in range(n): 
            cur += gain[i] 
            if cur > res: 
                res = cur 
        res = max(0, res)
        
        return res 
```
